```python
# set(text1) is the set of all the item types in text1, while len(text1) is the length of all the tokens in the text
# e.g. the ratio is somehow a measure of the lexical richness of the text, in this case for Moby Dick the ratio is 7,4%
len(set(text1))/len(text1)
```

```python
# this is how much the word "whale" is present in the text compared to all words
100 * text1.count("whale") / len(text1)
```

```python
def lexical_diversity(text):
    return len(set(text))/len(text)
def word_percentage(word,text):
    return 100 * text.count(word) / len(text)
```

```python
lexical_diversity(text2)
```

```python
word_percentage("whale",text1)
```

```python
fdist1 = FreqDist(text1)
print(fdist1)
fdist1.most_common(50)
```

```python
# Collocations are essentially just frequent bigrams, except that we want to pay more attention to the cases that involve
# rare words. In particular, we want to find bigrams that occur more often than we would expect based
# on the frequency of the individual words
text1.collocations()
```

# Counting other things than words


```python
# we can look at the distributions of word lenghts in a text by creating a FreqDist for the lenghts of each word
fdist = FreqDist(len(w) for w in text1)
fdist.most_common()
```

```python
# the word length most present in Moby Dick is three
fdist.max()
```

```python
# and a word long 3 chars is present ~20% of the times in the text
fdist.freq(3)
```
